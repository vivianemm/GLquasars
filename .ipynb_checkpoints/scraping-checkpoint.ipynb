```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
for table in soup.find_all('table'):
    print(table.prettify())
    print()
```

```python
tables = list(tables)
```
