```python
from astroquery.sdss import SDSS
```

```python
# psfMag: quasars and stars
# model and cmodel: galaxies (colors and photometry respectively)

query = """select top 1000  objID, type, z, modelMag_u, modelMag_g, modelMag_r, modelMag_i, modelMag_z, 
psfMag_u, psfMag_g, psfMag_r, psfMag_i, psfMag_z from specPhoto where zWarning = 0  ORDER BY NEWID()"""

result = SDSS.query_sql(query)
result = result.to_pandas()
```
